```python
# apply_model.py

import joblib
from sklearn.feature_extraction.text import CountVectorizer

# Load the saved model and vectorizer
model1 = joblib.load('multinomial_nb_model.joblib')
model2 = joblib.load('log_reg_model.joblib')
vectorizer = joblib.load('vectorizer.pkl')

# Function to preprocess and predict sentiment
def predict_sentiment(model, texts):
    # Transform the input text using the loaded vectorizer
    text_counts = vectorizer.transform(texts)

    # Predict using the loaded model
    predictions = model.predict(text_counts)

    return predictions

# Example usage
if __name__ == "__main__":
    new_texts = ["'Queen of Tears' successfully delivers the title as we all were in tears while watching each episode. This show had taken me on a emotional rollercoaster, with each passing episodes, from feeling heartbroken."]
    predictions = predict_sentiment(model1, new_texts)
    predictions2 =  predict_sentiment(model2, new_texts)
    print('NB Prediction:')
    for i in range(len(new_texts)):
      worded = ''
      if predictions[i] == 0:
        worded = 'negative'
      elif predictions[i] == 1:
        worded = 'positive'
      # else:
      #   worded = 'positive'

      print(new_texts[i])
      print(worded)


    # print(predictions)  # Output the predictions
```
